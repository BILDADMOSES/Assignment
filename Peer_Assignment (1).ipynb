```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("cardata.csv") #Reading the csv file
df.dropna(inplace = True) #Removing rows with empty values
nw_df = df[df["Selling_Price"] > 3.00 ]
nw_df
sp = nw_df["Selling_Price"] 
km = nw_df["Kms_Driven"]

plt.title("Relationship Between Km Driven and Selling Price")
plt.xlabel("Kms_Driven")
plt.ylabel("Selling_price")
plt.scatter(km,sp)
```
